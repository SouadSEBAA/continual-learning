```python
import pandas as pd



DATASET_PATH = './dataset/Combined.csv'
```

```python
df = pd.read_csv(DATASET_PATH, dtype={'Proto': 'str', 'sDSb': 'str', 'dDSb': 'str', 'Cause': 'str', 'State': 'str', 'Label': 'str', 'Attack Type': 'str', 'Attack Tool': 'str'})

df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0', 'Seq', 'Attack Tool', 'Label'], inplace=True)
df.head()
```

```python
df['sDSb'].unique()
```

```python
# df['sDSb'] = df['sDSb'].astype('category')
for col in df.columns:
    if df[col].isnull().sum() > (len(df) * 0.7):
        print("dropping", col, df[col].isnull().sum())
        df.drop(columns = [col], inplace=True)
df.head()
```

```python
# cleaning
print(df.dtypes)
for col in df.columns:
    if df.dtypes[col] == "object":
        print('object', col)
        df[col].fillna("NA", inplace=True)
    else:
        print('else', col)
        df[col].fillna(0, inplace=True)

df.info()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    if df.dtypes[col] == "object":
        df[col] = LabelEncoder().fit_transform(df[col])
        df[col] = df[col].astype('category')

df.info()
```

```python
attacks = df['Attack Type'].unique()
attacks
```
